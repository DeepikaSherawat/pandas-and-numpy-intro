```python
#add a new column
l=[]
for i in df["mpg"]:
    if i>20:
        l.append("high")
    else:
        l.append("low")

df["new_col"]=l
df   

    
```

```python
df["new_col"]=df["mpg"]>=22
df
```

```python

df.drop(["new_col"],axis=1,inplace=True)
#axis 0= row axis 1= column
#inplace= to save the value(to make the changes parmanent)
```
